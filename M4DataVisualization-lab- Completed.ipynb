# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

QUERY_CC = """
SELECT ConvertedComp FROM master
"""
df = pd.read_sql_query(QUERY_CC,conn)

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
plt.hist(df["ConvertedComp"], bins=6, edgecolor='black', alpha=0.7)
plt.xlabel('Compensation')
plt.ylabel('Frequency')
plt.title('Distribution')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY_BP = """
SELECT Age FROM master
"""
df_ag = pd.read_sql_query(QUERY_BP,conn)

df_ag.plot(kind='box', figsize=(10, 8))
plt.ylabel("Age")
plt.title("Age Box Plot")
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY_AW = """
SELECT Age, WorkWeekHrs FROM master
"""
df_aw = pd.read_sql_query(QUERY_AW,conn)

df_aw.plot(kind='scatter', x='WorkWeekHrs', y='Age', figsize=(10, 6), color='darkblue')
plt.title('Age Comparison against Work Week Hours')
plt.xlabel('Week Work Hours')
plt.ylabel('Age')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY_WCA= """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master
"""
df_wca = pd.read_sql_query(QUERY_WCA,conn)

x = df_wca['WorkWeekHrs']
y = df_wca['CodeRevHrs']
bubble_size = df_wca['Age']

plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=bubble_size * 20 + 5, alpha=0.5, c=bubble_size, cmap='viridis')


plt.xlabel('Work Hours per Week')
plt.ylabel('Code Review Hours')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')

cbar = plt.colorbar()
cbar.set_label('Age')

plt.grid(True)
plt.show()


```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# The table did not have the data available. Consequently, other fields were used to display the pie chart.

QUERY_RP= """
select count(Respondent) as Respondent_Total, Country from master 
GROUP BY Country
ORDER BY Respondent_Total DESC
LIMIT 5
"""
df_rp = pd.read_sql_query(QUERY_RP,conn)

plt.figure(figsize=(8, 8))
plt.pie(df_rp['Respondent_Total'], labels=df_rp['Country'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Respondent Total by Top 5 Countries')

plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY_SC= """
select Age, WorkWeekHrs, CodeRevHrs from master 
"""
df_sc = pd.read_sql_query(QUERY_SC,conn)
df_sc

filtered_df = df_sc[(df_sc['Age'] >= 30) & (df_sc['Age'] <= 35)]

median_values = filtered_df[['WorkWeekHrs', 'CodeRevHrs']].median()

fig, ax = plt.subplots(figsize=(8, 6))
median_values.plot(kind='bar', stacked=True, ax=ax)

plt.xlabel('Hour Categories')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Ages 30 to 35')

plt.xticks(rotation=0)
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY_LC= """
select Age, ConvertedComp from master where Age >= 45 AND Age <= 60
"""
df_lc = pd.read_sql_query(QUERY_LC,conn)

median_by_age = df_lc.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10, 6))
plt.plot(median_by_age.index, median_by_age.values,  linestyle='-', color='b', label='Median ConvertedComp')

plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp by Age 45 to 60')
plt.legend()

plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY_BC= """
select count(MainBranch) as Total, MainBranch from master group by MainBranch
"""
df_bc = pd.read_sql_query(QUERY_BC,conn)
df_bc

plt.figure(figsize=(10, 6))
plt.barh(df_bc['MainBranch'], df_bc['Total'], color='blue')

plt.xlabel('Total')
plt.ylabel('MainBranch')
plt.title('Total by MainBranch')

plt.show()
```
